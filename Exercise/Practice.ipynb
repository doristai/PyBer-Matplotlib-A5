````python
import matplotlib.pyplot as plt
````

````python
# Create the plot with ax.plt()
fig, ax = plt.subplots() # return a tuple containing a gigure and axes
ax.plot(x_axis, y_axis)
````

## Line Chart

````python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
````

````python
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x_axis, y_axis, marker ='D', color = 'green', linewidth = 2)
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.set_title('PyBer Fare by Month')
ax.set_ylim(0, 45)
ax.legend()
ax.grid()
````

## Bar Chart

````python
# Set the x-axis to a list of strings for each month.
months = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
fare = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
````

````python
plt.bar(months, fare, color = "green", label = 'Boston')
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.title('PyBar Fare by Month')
plt.legend()
````

````python
plt.barh(months, fare)
````

````python
plt.barh(months, fare, label = 'Boston', color = "purple")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()
plt.gca().invert_yaxis()
````

````python
fig, ax = plt.subplots()
ax.bar(months, fare)
````

````python
fig, ax = plt.subplots()
ax.barh(months, fare, color = 'cyan', alpha = 0.5, label = 'Chicago')
ax.invert_yaxis()
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
ax.legend()
````

## Scatter Plot

````python
plt.scatter(fare, months, label = "Chicago", marker = 'o', color = 'red')
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.xlim(0, 50)
plt.legend()
plt.gca().invert_yaxis()
````

## Bubble Chart

````python
# make the size of each marker equal to the data 
# floating-point decimal values of the y-axis by using s=y_axis
plt.scatter(months, fare, s=y_axis)
````

````python
fare_larger = []
for data in fare:
    fare_larger.append(data*3)
````

````python
plt.scatter(months, fare, s = fare_larger)
````

````python
# new_list = [expression for item in list if conditional]
plt.scatter(months, fare, s = [i*3 for i in fare])
````

````python
plt.scatter(fare, months, s = [i*5 for i in fare], color = 'skyblue', alpha = 0.8, edgecolor = 'black', linewidth = 2, label = 'Boston')
plt.gca().invert_yaxis()
plt.legend()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.xlim(0, 50)
plt.show()
````

## Pie Chart

````python
plt.pie(fare, labels = months)
plt.show
````

````python
# Increase the size of the pie chart 
plt.subplots(figsize=(8, 8))

# To explode a pie wedge
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# To add the percentage of each wedge on the pie chart, using autopct
# provide the format of one decimal place using '%.1f%%'
plt.pie(fare, explode = explode_values, labels = months, autopct='%.1f%%')
````

````python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(fare,
    explode = explode_values,
    colors = colors,
    labels = months,
    autopct = '%.1f%%', 
    shadow = True,
    startangle = 90,
    counterclock = 'counterclockwise')

plt.show()
````

### Create a Pie Chart Using the Object-Oriented Interface

````python
fig, ax = plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(fare, 
       labels = months, 
       explode = explode_values, 
       colors = colors,
       autopct = '%.1f%%')
plt.show()
````

## Add Error Bars

```plt.errorbar(x, y, yerr=<value>)```

````python
%matplotlib inline
````

````python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics as stats
````

````python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
````

````python
# Getting the SD of the value in y_axis
sd = stats.stdev(y_axis)
sd
````

````python
# Adding a cap to the rror bars that's equal to a scalar value
plt.errorbar(x_axis, y_axis, yerr = sd, capsize = 3)
````

````python
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=sd, capsize=3)
plt.show()
````

````python
plt.bar(x_axis, y_axis, yerr=sd, capsize=3)
````

````python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr = sd, capsize = 3)
plt.show()
````

````python
import numpy as np
plt.barh(x_axis, y_axis)

# Define the lower and upper limit of the x-axis 
# and the increment frequency $5
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
````

````python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
````

## Add Minor Ticks

```x.xaxis.set_minor_locator()```

````python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
````
